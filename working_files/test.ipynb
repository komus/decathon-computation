```python
def score_100m(secs_round):
    return round(25.4347 * pow((18 - secs_round), 1.81),0)

def score_long_jum(meters_jumped):
    return round(0.14354 * pow(((meters_jumped * 100) - 220), 1.4),0)

def score_shot_put(metres_thrown):
    return round(51.39 * pow((metres_thrown  - 1.5), 1.05),0)
    
def score_high_jump(metres_jumped):
    return round(0.8465 * pow(((metres_jumped * 100) - 75), 1.42),0)

def score_400m(sec_ran):
    return round(1.53775 * pow((82 - sec_ran), 1.81),0)

def score_110_hurdles(sec_ran):
    return 1000 if round(5.74352 * pow((28.5 - sec_ran), 1.92),0) > 1000 else round(5.74352 * pow((28.5 - sec_ran), 1.92),0)

def score_discus_throw(metres):
    return round(12.91 * pow((metres  - 4), 1.1),0)

def score_pole_vault(metres_jumped):
    return round(0.2797 * pow(((metres_jumped * 100) - 100), 1.35),0)

def score_javelin(metres):
    return round(10.14 * pow((metres  - 7), 1.08),0)

def score_1500m(mins_sec):
    split_time = mins_sec.split('.')
    secs = (60 * int(split_time[0])) + int(split_time[1]) + (int(split_time[-1])/1000)
    return 1000 if  round(0.03768 * pow((480 - secs), 1.85),0) > 1000 else round(0.03768 * pow((480 - secs), 1.85),0) 

def compute_score(data):
    m100 = data['100m'].apply(lambda x: score_100m(x))
    long_jump = data['long_jump'].apply(lambda x: score_long_jum(x))
    shot_pu = data['shot_put'].apply(lambda x: score_shot_put(x))
    high_jump = data['high_jump'].apply(lambda x: score_high_jump(x))
    m400 = data['400m'].apply(lambda x: score_400m(x))
    hurdles = data['110m_hurdles'].apply(lambda x: score_110_hurdles(x))
    discus = data['discus_throw'].apply(lambda x: score_discus_throw(x))
    pole = data['pole_vault'].apply(lambda x: score_pole_vault(x))
    m1500 =  data['1500m'].apply(lambda x: score_1500m(x))

    data['score'] = m100 + long_jump + shot_pu + high_jump + m400 + hurdles + discus + pole + m1500
    return data
```

```python
new_data = compute_score(data=data)
```

```python
new_data = new_data.sort_values(by=['score'], ascending=False)
```

```python
new_data.columns
```

```python
new_data["Rank_max"] = new_data["score"].rank(ascending = False, method = 'max').astype(int)
new_data["Rank_min"] = new_data["score"].rank(ascending = False, method = 'min').astype(int)
```

```python
def get_ranking(min_rank, max_rank):
    if min_rank == max_rank:
        return min_rank
    else:
        return f'{min_rank} - {max_rank}'
```

```python
new_data["placement"] = new_data.apply(lambda dat: get_ranking(dat['Rank_min'], dat['Rank_max']), axis = 1)
```
